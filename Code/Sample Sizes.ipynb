```python
import math
import sys

import numpy as np
import matplotlib.pyplot as plt

from assertion_audit_utils import TestNonnegMean
```

```python
def compute_asn(alpha, margin):
    s_w = .5+margin
    s_l = 1 - s_w
    z_w = math.log(2 * s_w)
    z_l = math.log(2 - 2 * s_w)

    return math.ceil((math.log(1/alpha) + (z_w / 2.0)) / (s_w*z_w + s_l*z_l))
```

```python
def do_sprt_trials(ntrials, margin, alpha):
    avg = 0
    for i in range(ntrials): 
        val = TestNonnegMean.wald_sprt_sample_size(np.infty, margin, .002, alpha=alpha)
        print(val)
        avg += val
        
    return avg/ntrials

do_sprt_trials(10, .4, .05)
```

```python

ntrials = 10
alpha = .05

km = []
asns = []
sprt = []
for i in range(1, 50):
    sys.stdout.write('{}\r'.format(i))
    margin = i/100
    km.append(TestNonnegMean.kaplan_martingale_sample_size(np.infty, margin, .002, alpha=alpha))
    asns.append(compute_asn(alpha, margin))
    sprt.append(do_sprt_trials(ntrials, margin, alpha))
print(km)
print(sprt)
plt.plot(km)
plt.plot(asns)
plt.plot(sprt)
plt.yscale('log')

plt.show()
```
